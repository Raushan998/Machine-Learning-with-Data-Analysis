```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import nltk
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data=pd.read_csv('/home/raushan/Desktop/Machine_Learning/projects and Assignments/stocknews/Combined_News_DJIA.csv')
```

```python
import re
nltk.download('stopwords')
```

```python
train = data[data['Date'] < '2015-01-01']
test = data[data['Date'] > '2014-12-31']
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
trainheadlines=[]
for row in range(0,len(train.index)):
    trainheadlines.append(" ".join(str(x) for x in train.iloc[row,2:27]))
```

```python
corpus=[]
for i in range(0, len(trainheadlines)):
        review = re.sub('[^a-zA-Z]', ' ', trainheadlines[i])
        review = review.lower()
        review = review.split()
        ps = PorterStemmer()
        review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
        review = ' '.join(review)
        corpus.append(review)
```

```python
trainheadlines1=[]
for row in range(0,len(test.index)):
    trainheadlines1.append(" ".join(str(x) for x in test.iloc[row,2:27]))
```

```python
corpus1=[]
for i in range(0, len(trainheadlines1)):
        review = re.sub('[^a-zA-Z]', ' ', trainheadlines[i])
        review = review.lower()
        review = review.split()
        ps = PorterStemmer()
        review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
        review = ' '.join(review)
        corpus1.append(review)
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
```

```python
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression(random_state=0)
regressor.fit(X,train['Label'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X_test = cv.fit_transform(corpus1).toarray()
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(test['Label'],y_pred)
```

```python
confusion_matrix(y_pred,test["Label"])
```

```python
from sklearn.naive_bayes import GaussianNB
regressor1=GaussianNB()
```

```python
regressor1.fit(X,train['Label'])
y_pred=regressor1.predict(X_test)
```

```python
accuracy_score(test['Label'],y_pred)
```

```python
from sklearn.naive_bayes import MultinomialNB
regressor2=MultinomialNB(alpha=0.01)
regressor2.fit(X,train['Label'])
y_pred=regressor2.predict(X_test)
```

```python
accuracy_score(test['Label'],y_pred)
```

```python
from sklearn.svm import SVC
regressor3=SVC(kernel='rbf',random_state=0)
regressor3.fit(X,train['Label'])
y_pred=regressor3.predict(X_test)
```

```python
accuracy_score(test['Label'],y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
regressor4=DecisionTreeClassifier(criterion='entropy',random_state=0)
regressor4.fit(X,train['Label'])
y_pred=regressor4.predict(X_test)
```

```python
accuracy_score(test['Label'],y_pred)
```
